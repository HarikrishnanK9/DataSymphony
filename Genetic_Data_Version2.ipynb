```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dec=DecisionTreeClassifier()
dec.fit(x_train,y_train)
y_pred=dec.predict(x_test)
y_pred
```

```python
df.Ancestry.unique()
```

```python
from sklearn import tree
plt.figure(figsize=(15,15))
tree.plot_tree(dec,feature_names=["PC1","PC2","PC3","PC4","PC5","PC6","PC7","PC8","PC9","PC10"],class_names=['African', 'European', 'EastAsian', 'Oceanian', 'NativeAmerican','Unknown', 'Mexican', 'AfricanAmerican'],filled=True)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print("Accuracy Score is :",accuracy_score(y_test,y_pred))
print("Classification Report:")
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
result=confusion_matrix(y_test,y_pred)
print(result)
labels=['African', 'European', 'EastAsian', 'Oceanian', 'NativeAmerican','Unknown', 'Mexican', 'AfricanAmerican']
cmd=ConfusionMatrixDisplay(result,display_labels=labels)
cmd.plot()
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=10, random_state=42)
rfc.fit(x_train,y_train)
y_pred1=rfc.predict(x_test)
y_pred1
```

```python
# feature_names=["PC1","PC2","PC3","PC4","PC5","PC6","PC7","PC8","PC9","PC10"],class_names=['African', 'European', 'EastAsian', 'Oceanian', 'NativeAmerican','Unknown', 'Mexican', 'AfricanAmerican'],filled=True)
```

```python
# Plot the decision trees in the random forest with a limited depth (e.g., max_depth=3)
plt.figure(figsize=(20, 10))
for i in range(len(rfc.estimators_)):
    tree.plot_tree(rfc.estimators_[i], filled=True, feature_names=["PC1","PC2","PC3","PC4","PC5","PC6","PC7","PC8","PC9","PC10"], class_names=['African', 'European', 'EastAsian','Oceanian','NativeAmerican','Unknown', 'Mexican', 'AfricanAmerican'])
plt.show()
```

```python
from sklearn.metrics import accuracy_score,classification_report
print("Accuracy Score is :",accuracy_score(y_test,y_pred1))
print("Classification Report:")
print(classification_report(y_test,y_pred1))
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
result=confusion_matrix(y_test,y_pred1)
print(result)
labels=['African', 'European', 'EastAsian', 'Oceanian', 'NativeAmerican','Unknown', 'Mexican', 'AfricanAmerican']
cmd=ConfusionMatrixDisplay(result,display_labels=labels)
cmd.plot()
```
